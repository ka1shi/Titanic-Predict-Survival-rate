```python
#Replacing missing age value with median age of respective title of person

data_train['Age'].fillna(-1, inplace=True)

median_age = dict()
for title in data_train['Name']:
    m = data_train.Age[(data_train["Age"] != -1) & (data_train['Name'] == title)].median()
    median_age[title] = m
    
for index, row in data_train.iterrows():
    if row['Age'] == -1:
        data_train.loc[index, 'Age'] = median_age[row['Name']]

data_train.apply(lambda x : sum(x.isnull()), axis=0)
```

```python
#Replacing missing age value with median age of respective title of person

data_test['Age'].fillna(-1, inplace=True)

median_age = dict()
for title in data_test['Name']:
    m = data_test.Age[(data_test["Age"] != -1) & (data_test['Name'] == title)].median()
    median_age[title] = m

for index, row in data_test.iterrows():
    if row['Age'] == -1:
        #print(index)
        data_test.loc[index, 'Age'] = median_age[row['Name']]

data_test.apply(lambda x : sum(x.isnull()), axis=0)
```

```python
# there is still an age ie is missing, lets check it
null_data = data_test[data_test.isnull().any(axis=1)]
print(null_data)
```

```python
# because title 'Ms' has one age and that was missing 
# we can replace missing age here with title 'Mrs' as an exception 
# can not delete the row as is part test data

temp = median_age['Mrs']
print(temp)
data_test['Age'].fillna(temp, inplace=True)
```
